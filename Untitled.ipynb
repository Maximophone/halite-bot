```python
def plot_squares(locs,height,width,res=2):
    im = np.zeros((height*res,width*res))
    for loc in locs:
        im[loc.x*res:loc.x*res+res,loc.y*res:loc.y*res+res]=1
    plt.imshow(im)
```

```python
def plot_quantity(gameMap,quantity,res=2,function=None):
    if function is None:
        function = lambda x:x
    im = np.zeros((gameMap.height*res,gameMap.width*res))
    for y in range(gameMap.height):
        for x in range(gameMap.width):
            loc = Location(x,y)
            im[loc.x*res:loc.x*res+res,loc.y*res:loc.y*res+res]=function(getattr(gameMap.getSite(loc),quantity))
    plt.imshow(im)
```

```python
def weightedChoice(choices,weights):
    interval = sum(weights)
    x = random.random()*interval
    cum_weights = 0
    for i,w in enumerate(weights):
        cum_weights += w
        if x <= cum_weights:
            return choices[i]
```

```python
def dist_frontier(loc,myID,gameMap,to_search=None,searched=None):
    if to_search is None:
        to_search = deque([loc])
    if searched is None:
        searched = {(x,y):0 for x in range(gameMap.width) for y in range(gameMap.height)}
    while len(to_search)>0:
        selected = to_search.popleft()
        searched[(selected.x,selected.y)] = 1
        for d in CARDINALS:
            new_loc = gameMap.getLocation(selected,d)
            if not searched[(new_loc.x,new_loc.y)]:
                to_search.append(new_loc)
        site = gameMap.getSite(selected)
        if site.owner != myID:
            site.dist_frontier = -1
        
    
```

```python
def find_frontier(myID,gameMap):
    frontier = []
    for y in range(gameMap.height):
        for x in range(gameMap.width):
            loc = Location(x,y)
            site = gameMap.getSite(loc)
            if site.owner == myID:
                continue
            for d in CARDINALS:
                if gameMap.getSite(loc,d).owner != myID:
                    continue
                else:
                    frontier.append(loc)
                    break
    return frontier
                
```

```python
def dist_frontier(frontier,myID,gameMap,dist=-1):
    if not frontier:
        return
    new_frontier = []
    for loc in frontier:
        for c in CARDINALS:
            new_loc = gameMap.getLocation(loc,c)
            site = gameMap.getSite(new_loc)
            if site.owner!=myID or site.dist_frontier is not None:
                continue
            site.dist_frontier = dist+1
            new_frontier.append(new_loc)
    dist_frontier(new_frontier,myID,gameMap,dist=dist+1)
```

```python
def mapDistFrontier(myID,gameMap):
    frontier = find_frontier(myID,gameMap)
    dist_frontier(frontier,myID,gameMap)
```

```python
def custom_min(x):
    v_min = 999999
    a_min = None
    for v,a in x:
        if v<v_min:
            v_min = v
            a_min = a
    return a_min
```

```python
def mapFrontierDirections(gameMap):
    for y in range(gameMap.height):
        for x in range(gameMap.width):
            loc = Location(x,y)
            site = gameMap.getSite(loc)
            if site.dist_frontier is None or site.dist_frontier <= 0:
                site.frontierDir = None
            else:
                fmoves = [(gameMap.getSite(loc,d).dist_frontier,d) for d in CARDINALS]
                random.shuffle(fmoves)  
                site.frontierDir = custom_min(fmoves)
```
